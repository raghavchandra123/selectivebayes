```python
import pandas as pd
import numpy as np
from sklearn.cluster import SpectralClustering
pd.set_option('max_colwidth', 400)
pd.set_option("display.max_rows",500)
```

```python
pockets = pd.read_csv("Report_PoSSuM.txt",sep="|")
```

```python
diffreses = []
for pocket in pockets["Aligned residues"]:
    aligned = pocket.split(",")
    diffres=set()
    for res in aligned:
        truncres = res[2:res.find("-")+1]+res[res.find("-")+3]
        if truncres[-1]!=truncres[0]:
            diffres.add(truncres)
    diffreses.append(diffres)
pockets["Different residues"]=diffreses
pockets.head()
```

```python
pockets.drop(columns=["Chain ID","EC No.","CATH code","SCOPe code"])
```

```python
sortedpocks = pockets.sort_values(by="Cosine value",ascending=False)
uniquepocks = sortedpocks.drop_duplicates(subset="UniProt ID")
```

```python
uniquepocks[["Protein Name","Different residues"]]
```

```python
uniquesets=[]
for i in range(len(uniquepocks)):
    setunit = uniquepocks.iloc[i]["Different residues"]
    for j in range(len(uniquesets)):
        if setunit.issuperset(uniquesets[j]):
            break
        if setunit.issubset(uniquesets[j]):
            if len(uniquesets[j])>len(setunit):
                uniquesets[j]=setunit
            break
    else:
        uniquesets.append(setunit)
print(uniquesets)
print(len(uniquesets))

```

```python
simmatrix = np.zeros((len(uniquepocks),len(uniquepocks)))
for i in range(len(uniquepocks)):
    for j in range(len(uniquepocks)):
        simmatrix[i,j]=len(uniquepocks.iloc[i]["Different residues"].intersection(uniquepocks.iloc[j]["Different residues"]))/len(uniquepocks.iloc[i]["Different residues"].union(uniquepocks.iloc[j]["Different residues"]))
simmatrix
```

```python
nodifferent = []
for i in uniquepocks["Different residues"]:
    nodifferent.append(len(i))
uniquepocks["No differences"]=nodifferent
```

```python
uniquepocks["frac differences"]=uniquepocks["No differences"]/uniquepocks["Aligned length"]
```

```python
clustering = SpectralClustering(n_clusters = 20,affinity="precomputed").fit(simmatrix)
uniquepocks["cluster"]=clustering.labels_
uniquepocks[["Cosine value","Aligned length","Protein Name","Different residues","cluster","No differences","frac differences"]]
```

```python
interesting = uniquepocks.loc[uniquepocks["Different residues"].isin(uniquesets)]
print(len(interesting))
interesting.head(200)
```
